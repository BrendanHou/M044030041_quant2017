# Q1. 畫出 diamond shape

```python
def pyramid(n):
    for i in range(n):
        print(" "*(n-i) + "*"*(2*i+1))
def reverse_pyramid(n):
    for i in range(n, -1, -1):
        print(" "*(n-i) + "*"*(2*i+1))
```

```python
pyramid(5)
reverse_pyramid(5)
```

# Q2. 刪除重複性的資料。

```python
import pandas as pd
```

```python
# 取得公司資料
def get_companies(ex = "NASDAQ"):
    template = "http://www.nasdaq.com/screening/companies-by-industry.aspx?exchange={}&render=download"
    url = template.format(ex)
    return pd.read_csv(url)
```

```python
df = get_companies()
```

```python
df.head()
```

```python
symbols_01 = df[(df['MarketCap'] > 3e8) & (df['MarketCap'] < 4e8)]['Symbol'].tolist()
```

```python
df['LastSale'] = df['LastSale'].str.replace('n/a', '0.0').astype(float)
```

```python
symbols_02 = df[(df['LastSale'] > 3) & (df['LastSale'] < 5)]['Symbol'].tolist()
```

```python
with open('symbols01.txt', 'w') as f:
    for symbol in symbols_01:
        f.write(symbol.strip().upper() + '\n')
```

```python
with open('symbols02.txt', 'w') as f:
    for symbol in symbols_02:
        f.write(symbol.strip().upper() + '\n')
```

```python
f = open('symbols01.txt')
code_01 = f.read()
code_01 = code_01.split()
print(code_01)
f.close()
```

```python
f = open('symbols02.txt')
code_02 = f.read()
code_02 = code_02.split()
print(code_02)
f.close()
```

```python
code_all = list(set(code_01 + code_02))
code_all
```

```python
with open('symbols03.txt', 'w') as f:
    for symbol in code_all:
        f.write(symbol.strip().upper() + '\n')
```

```python
f = open('symbols03.txt')
symbols_03 = f.read()
symbols_03 = symbols_03.split()
print(symbols_03)
f.close()
```
